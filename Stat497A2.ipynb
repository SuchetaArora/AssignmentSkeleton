```python
#insert module to create static visualization
import matplotlib.pyplot as plt

#Create function that draws a tennis double court
def create_tennis_court(filename="tennis_court_drawing.png"):
    # Court dimensions in ft
    court_length_ft = 78
    court_width_ft = 36
    service_lane_length_ft = 18  
    service_lane_width_ft = 27 
    middle_line_length_ft = 42
    double_alley_width_ft = 4.5  

    # Create a Matplotlib figure and axis (appear like a graph)
    fig, ax = plt.subplots(figsize=(15, 7))  # Increase the figure size

    # Adjust plot limits with some extra space to accurately represent the nonplaying surface
    ax.set_xlim(-21, court_length_ft + 21)  #120ft
    ax.set_ylim(-12, court_width_ft + 12)   #60ft

    # Set the background color outside the court to brown
    ax.set_facecolor('#37ab6c')

    # Draw court outline with white outlines and forest green fill
    court_outline = plt.Rectangle((0, 0), court_length_ft, court_width_ft, fill=True, facecolor="#3776ab", edgecolor="white", lw=3)
    ax.add_patch(court_outline)

    # Draw service lanes with the gap from 4.5ft away from the width border line
    service_lane_x1 = 4.5  
    service_lane_x2 = service_lane_x1 + service_lane_length_ft
    ax.plot([0, court_length_ft], [service_lane_x1, service_lane_x1], color="whitesmoke", lw=3)

    # Draw the service lines on both sides of the net
    service_line_x1 = 18  # 18 feet from the start
    service_line_x2 = court_length_ft - 18  # symmetric side from the net
    service_line_y1 = 4.5  # 4.5 feet in height
    service_line_y2 = 31.5  # Adjusted to 27 feet in height
    ax.plot([service_line_x1, service_line_x1], [service_line_y1, service_line_y2], color="white", lw=3)
    ax.plot([service_line_x2, service_line_x2], [service_line_y1, service_line_y2], color="white", lw=3)

    # Draw center service line
    center_service_line_x1 = 18  # 18 feet from the start
    center_service_line_x2 = 60  # 60 feet
    center_service_line_y1 = 18  # Starting height
    center_service_line_y2 = 18  # Ending height
    ax.plot([center_service_line_x1, center_service_line_x2], [center_service_line_y1, center_service_line_y2], color="white", lw=3)

    # Draw a line representing the net
    net_x = court_length_ft / 2
    net_y1 = -3  # Adjusted starting y-coordinate
    net_y2 = 39  # Adjusted ending y-coordinate
    ax.plot([net_x, net_x], [net_y1, net_y2], color="whitesmoke", lw=1)
    
    
    ax.plot([net_x, net_x], [0, court_width_ft], color="whitesmoke", lw=1)
    
    # Draw net posts 
    net_posts = [(39, -3), (39, 1.5), (39, 34.5), (39, 39)]
    for post_x, post_y in net_posts:
        ax.add_patch(plt.Circle((post_x, post_y), radius=0.3, color="black", lw=3))

    # Draw shorter center marks
    center_mark_x1 = 0  # x-coordinate for the left center mark
    center_mark_x2 = 78  # x-coordinate for the right center mark
    center_mark_y = 18  # The height of the center marks
    mark_length = 0.33  # Real length of mark is 4inch=0.33ft

    # Left center mark
    ax.plot([center_mark_x1 - mark_length / 2, center_mark_x1 + mark_length / 2], [center_mark_y, center_mark_y], color="white", lw=3.5)

    # Right center mark
    ax.plot([center_mark_x2 - mark_length / 2, center_mark_x2 + mark_length / 2], [center_mark_y, center_mark_y], color="white", lw=3.5)

    # Draw outer lines
    outer_line_y1 = double_alley_width_ft
    outer_line_y2 = court_width_ft - double_alley_width_ft
    ax.plot([0, court_length_ft], [outer_line_y1, outer_line_y1], color="white", lw=3)
    ax.plot([0, court_length_ft], [outer_line_y2, outer_line_y2], color="white", lw=3)

    # Draw double alleys
    left_double_alley_x = 0
    right_double_alley_x = court_length_ft
    ax.plot([left_double_alley_x, right_double_alley_x], [outer_line_y1, outer_line_y1], color="white", lw=3)
    ax.plot([left_double_alley_x, right_double_alley_x], [outer_line_y2, outer_line_y2], color="white", lw=3)

    # Add X-axis and Y-axis labels
    ax.set_xlabel('Length (ft)')
    ax.set_ylabel('Width (ft)')

    # Set the X-axis and Y-axis ticks and labels
    x_ticks = [0, service_line_x1, court_length_ft/2, service_line_x2, court_length_ft]
    y_ticks = [0, service_lane_x1, 18,36-4.5, court_width_ft]
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    ax.set_xticklabels([f"{val} ft" for val in x_ticks])
    ax.set_yticklabels([f"{val} ft" for val in y_ticks])

    # Save the image as a PNG file
    plt.savefig(filename, bbox_inches='tight', pad_inches=0, dpi=300)

# Call the function to draw and save the tennis court
create_tennis_court("tennis_court_drawing.png")
```
